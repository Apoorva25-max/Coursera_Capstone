```python
X = df[['ADDRTYPE','WEATHER', 'ROADCOND']].values
X
```

```python
df = df[['SEVERITYCODE']].replace(1,"Low")
df = df[['SEVERITYCODE']].replace(2, "High")
```

```python
y = df[['SEVERITYCODE']].values
y
```

```python
from sklearn import preprocessing
```

```python
df_addr = preprocessing.LabelEncoder()
df_addr.fit(('Intersection', 'Block', 'Alley'))
X[:,0] = df_addr.transform(X[:,0])
```

```python
df_weather = preprocessing.LabelEncoder()
df_weather.fit(('Overcast', 'Raining', 'Clear', 'Unknown', 'Other', 'Snowing', 'Fog/Smog/Smoke', 'Sleet/Hail/Freezing Rain', 'Blowing Sand/Dirt','Severe Crosswind', 'Partly Cloudy'))
X[:,1] = df_weather.transform(X[:,1])
```

```python
df_road = preprocessing.LabelEncoder()
df_road.fit(('Wet', 'Dry', 'Unknown', 'Snow/Slush', 'Ice', 'Other','Sand/Mud/Dirt', 'Standing Water', 'Oil'))
X[:,2] = df_road.transform(X[:,2])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X
```

```python
y
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X,y,test_size = 0.3, random_state=3)
print(X_trainset.shape)
print(y_trainset.shape)
print(X_testset.shape)
print(y_testset.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
severityTree = DecisionTreeClassifier()
severityTree.fit(X_trainset, y_trainset)
```

```python
predTree = severityTree.predict(X_testset)
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("Decision Tree's Accuracy:", metrics.accuracy_score(y_testset, predTree))
```
